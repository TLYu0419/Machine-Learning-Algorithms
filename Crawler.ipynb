# 載入套件

```python
import requests
import pandas as pd
import re
from requests_html import HTML
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

# 抓取網站有哪些大標題與期連結

```python
url = 'https://blog.datasciencedojo.com/machine-learning-algorithms/'
def function006(url):
    driver.get(url)
    soup = BeautifulSoup(driver.page_source)
    ids = []
    for i in soup.select('h2')[1:-1]:
        ids.append(i['id'])
    dess = []
    for i in soup.select('p')[4:-13]:
        dess.append(i.text)
    srcs = []
    for i in soup.select('iframe')[:-2]:
        srcs.append(i['src'])
    return pd.DataFrame({'ids':ids,
                         'dess':dess,
                         'srcs':srcs})
df = function006(url)
```

# 抓取連接中的算法資訊

```python
def function007(url):
    driver.get(url)
    soup = BeautifulSoup(driver.page_source)
    title = []
    for i in soup.select('h3.h5p-panel-title'):
        title.append(i.text)
        
    content = []
    for i in soup.select('div.h5p-panel-content'):
        tmp = ''
        for j in i.select('p'):
            tmp = tmp + j.text
        content.append(tmp)

    links = []
    for i in soup.select('div.h5p-panel-content'):
        tmp = ''
        for j in i.select('ul li a'):
            tmp = tmp + '[' + j.text + ']' + '(' + j['href'] + ')'
        links.append(tmp)
    
    return pd.DataFrame({'title':title,
                         'content':content,
                         'links':links})
```

```python
df2 = pd.DataFrame()
for id,  src in zip(df.ids, df.srcs):
    ndf = function007(src)
    ndf['ids'] = id
    df2 = pd.concat([df2, ndf],ignore_index = True)
df2
```
